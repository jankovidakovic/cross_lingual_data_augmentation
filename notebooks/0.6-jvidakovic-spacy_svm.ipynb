```python
import spacy
nlp = spacy.load("en_core_web_trf", disable=["parser", "attribute-ruler", "ner"])
nlp.add_pipe("sentencizer")
```

```python
import pandas as pd

df = pd.read_csv("../data/dummy.csv")
df.head()
```

```python
def yield_column(df, column):
    yield from df.loc[:, column].values
```

```python
[i for i in yield_column(df, "text")]
```

```python
from itertools import islice
from typing import Iterable

texts = (x for x in yield_column(df, "text"))
def process(nlp, texts: Iterable[str], n_sentences: int):
    with nlp.select_pipes(enable=["transformer", "tagger", "attribute_ruler", "sentencizer", "lemmatizer"]):
        for doc in nlp.pipe(texts, n_process=-1):
            sentences = (sent for sent in doc.sents)
            sentences = islice(sentences, 0, n_sentences)
            # now we would need a concat map
            # yield (token.lemma_ for token in sentences)
            yield (token.lemma_ for sentence in sentences for token in sentence)
```

```python
from tqdm import tqdm

result = list(map(list, tqdm(process(nlp, texts, 25), desc="Pipeline", total=10)))
result
```
